```python
import pandas as pd
```

```python
authors = {
    'author_id':[1,2,3,4,5],
    'author_name':['Тургенев','Чехов','Островский','Пушкин','Лермонтов']
}
authors = pd.DataFrame(authors)
authors
```

```python
books={
    'author_id':[1,1,1,2,2,3,3,4,4,5],
    'book_title':['Отцы и дети','Рудин','Дворянское гнездо','Толстый и тонкий','Дама с собачкой','Гроза',
                  'Таланты и поклонники','Евгений Онегин','Дубровский','Герой нашего времени'],
    'price':[450,300,350,500,450,370,290,400,430,575]
}
books = pd.DataFrame(books)
books
```

```python
authors_price = pd.merge(authors,books)
authors_price
```

```python
top_5 = authors_price.nlargest(5,'price')
top_5
```

```python
stat_1 = authors_price.groupby('author_name').agg({'price':"min"})
stat_2 = authors_price.groupby('author_name').agg({'price':"max"})
stat_3 = authors_price.groupby('author_name').agg({'price':"mean"})

stat_1 = stat_1.rename(columns={'price':'min_price'})
stat_2 = stat_2.rename(columns={'price':'max_price'})
stat_3 = stat_3.rename(columns={'price':'mean_price'})

authors_stat= pd.concat([stat_1,stat_2,stat_3],axis=1)

authors_stat
```
