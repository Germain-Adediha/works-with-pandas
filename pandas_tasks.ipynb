```python
import pandas as pd
```

```python
authors = {
    'author_id':[1,2,3,4,5],
    'author_name':['Тургенев','Чехов','Островский','Пушкин','Лермонтов']
}
authors = pd.DataFrame(authors)
authors
```

```python
books={
    'author_id':[1,1,1,2,2,3,3,4,4,5],
    'book_title':['Отцы и дети','Рудин','Дворянское гнездо','Толстый и тонкий','Дама с собачкой','Гроза',
                  'Таланты и поклонники','Евгений Онегин','Дубровский','Герой нашего времени'],
    'price':[450,300,350,500,450,370,290,400,430,575]
}
books = pd.DataFrame(books)
books
```

```python
authors_price = pd.merge(authors,books)
authors_price
```

```python
top_5 = authors_price.nlargest(5,'price')
top_5
```
